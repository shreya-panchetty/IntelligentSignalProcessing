#### Task 1- Application for detecting and tracking moving cars from a camera recording

###### Traffic_Laramie_1 video file

```python
#importing required modules
import cv2

#initialising videoCapture object and reading the frames from the input file
video_file=cv2.VideoCapture('Traffic_Laramie_1.mp4')


#Verifing that the video has properly opened
if not video_file.isOpened():
    #printing the error message incase the video failed to open
    print("Error in opening the video_file.")
    #exiting the code
    exit(0)

#initialising a k-neraest neighbours background subtraction instance.
bgSubtract_instance=cv2.createBackgroundSubtractorKNN(detectShadows=False)

#reading the video_file completely
while True:
    #initialising frame-by-frame rate
    check_rate,frame_rate=video_file.read()
    
    #stopping the loop constant if the check_rate is false
    if check_rate==False:
        break;

    #initialising and calculating the foreground mask
    #difference between frame_rate and calculated bgSubtract_instance
    foreground_mask=bgSubtract_instance.apply(frame_rate)
    
    #dilating the foreground_mask
    foreground_mask=cv2.dilate(foreground_mask,None,18)
    #eroding the foreground_mask to get the detection boxes
    foreground_mask=cv2.erode(foreground_mask,None,10)
    
    #initialising the threshold_frameRate
    threshold_frameRate=cv2.threshold(foreground_mask,100,255,cv2.THRESH_BINARY)[1]
    
    #initialising contoursRate to identify all the contours present
    (contoursRate,_)=cv2.findContours(threshold_frameRate,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

    #looping over contoursRate value 
    for cr in contoursRate:
        # initialising the bounding rectangle around the objects
        (x_val, y_val, width_val, height_val)=cv2.boundingRect(cr)
        #filtering the contours identifying human shapes
        if (height_val/width_val > 2.5)  | (cv2.contourArea(cr) < 1500):
            continue

        #Drawing the bounding boxes around the identified cars passing
        cv2.rectangle(frame_rate, (x_val, y_val), (x_val+width_val, y_val+height_val), (0,255,255), 1)
    

    #Displaying the various frames on screen
    if check_rate == True:
        #displaying the webcam video_file
        cv2.imshow('Webcam video_file', frame_rate)
        #displaying the threshold frame video_file
        cv2.imshow('Threshold frame video_file', threshold_frameRate)
        #displaying the foreground_mask video_file
        cv2.imshow('Foreground mask video', foreground_mask)
        #break playing the video if the key "q" is pressed or video is played completely 
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

#Releasing the video_file object
video_file.release()

#destroying all the windows
cv2.destroyAllWindows()
```

#### Task 2- Counting the number of cars going to city's downtown during peak hours

###### Traffic_Laramie_1 video file

```python
#importing the required modules
from datetime import datetime
import cv2

#initialising videoCapture object and reading the frames from the input file
video_file=cv2.VideoCapture('Traffic_Laramie_1.mp4')

#Verifing that the video has properly opened
if not video_file.isOpened():
    #printing the error message incase the video failed to open
    print("Error in opening the video_file.")
    #exiting the code
    exit(0)

#initialising a k-neraest neighbours background subtraction instance.
bgSubtract_instance=cv2.createBackgroundSubtractorKNN(detectShadows=False)

#storing the bound boxes from previous frame
previous_BoundBox=[]

#initialising values used to match the current frame's bounding boxes to those from the previous frame
#X-axis 
Xposition_value=10
#Y-axis
Yposition_value=10
#horizontal size
heightPosition_value=10
#vertical size
widthPosition_value=10

#initialising threshold value that determining the 
threshold_value=200

#initialising threshold value detrmining the number of frames the boundboxes have tracked
num_frames_threshold=10

#number of vehicles per minute 
cars_passing_perMinute=0

#start time for video
startTime_video=datetime.now()

#initialising the counter for vehicles passing
vehicles_counter=0

#class function of bounding boxes
class BoundBoxesMotion:
    def __init__(self,x,y,w,h):
        self.x=x
        self.y=y
        self.w=w
        self.h=h
        
        #initialising the dierction of the movemnt
        self.direction=''
        #checking whether the box/vehicle has been counted
        self.counted_box=False
        #counting the no. of frames the boundingBoxes have tracked
        self.noOfFrames=0
        
#reading the video_file completely
while True:
    #initialising frame-by-frame rate
    check_rate,frame_rate=video_file.read()
    
    #stopping the loop constant if the check_rate is false
    if check_rate==False:
        break;

    #initialising and calculating the foreground mask
    #difference between frame_rate and calculated bgSubtract_instance
    foreground_mask=bgSubtract_instance.apply(frame_rate)
    
    #dilating the foreground_mask
    foreground_mask=cv2.dilate(foreground_mask,None,18)
    #eroding the foreground_mask to get the detection boxes
    foreground_mask=cv2.erode(foreground_mask,None,10)
    
    #initialising threshold_frameRate
    threshold_frameRate=cv2.threshold(foreground_mask,100,255,cv2.THRESH_BINARY)[1]
    
    #initialising contoursRate to identify all the contours present
    (contoursRate,_)=cv2.findContours(threshold_frameRate,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    
    #storing the number of boundBoxes presnet in the current frame o fthe video
    current_BoundBox=[]
    
    #looping over contoursRate value 
    for cr in contoursRate:
        # nitialising the bounding rectangle around the objects
        (x, y, w, h)=cv2.boundingRect(cr)
        #filtering the contours identifying human shapes
        if (h/w > 2)  | (cv2.contourArea(cr) < 1500):
            continue
        
        #bounding boxes rectangle size values
        cv2.rectangle(frame_rate,(x,y),(x+w,y+h),(0,255,255),1)
        #creating and adding boundingBox depending on the motion of the vehicle
        current_BoundBox.append(BoundBoxesMotion(x,y,w,h))
        
    #Looping over the current_BoundBox array  
    for currentBox in current_BoundBox:
        #if the boundBox hasnt been detected before
        found_box=False
        
        #looping over the previous_BoundBox array
        for previousBox in previous_BoundBox:
            #calculating the distance between current and previous boundboxes
            X_value=currentBox.x-previousBox.x
            Y_value=currentBox.y-previousBox.y
            
            #calculating the difference between framerate of current and previous
            W_value = currentBox.w - previousBox.w
            H_value = currentBox.h - previousBox.h
            
            #checking if the condition reamins false
            if (abs(X_value) > Xposition_value) | (abs(Y_value) > Yposition_value) \
                | (abs(W_value) > widthPosition_value) |  (abs(H_value) > heightPosition_value):
                #continuing the code as the currentBox and previousBox are apart
                continue
                
            #if the condition is true
            found_box = True
            
            #Using putText function to labelling the box, direction and no.of frames of the vehicle 
            cv2.putText(frame_rate, f'direction:{previousBox.direction}, frames:{previousBox.noOfFrames}', 
                        (currentBox.x, currentBox.y), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,255,255), 1, cv2.LINE_AA)
                       

            #Calculating the direction based on X_value
            #if value of x is more than 0
            if X_value > 0:
                #initialising direction is right
                currentBox.direction = 'right'
            #if value of X is less than 0
            elif X_value < 0:
                #initialising direction is left
                currentBox.direction = 'left'
                #updating noOfframes of currentBox 
                currentBox.noOfFrames = previousBox.noOfFrames + 1
            else:
                #initialising direction is straight
                currentBox.direction = 'straight'

            #tracking the number of counted boxes   
            currentBox.counted_box = previousBox.counted_box
                
                
            #condition to check if vehicle goes out of frame of the video_file 
            if (currentBox.x < threshold_value) and (currentBox.noOfFrames > num_frames_threshold) \
                and (currentBox.direction == 'left') and (currentBox.counted_box== False):
                #Updating the vehicles_counter
                vehicles_counter+=1
                #checking currentBox has correctly counted the box
                currentBox.counted_box = True
                
        #Checking id the previousBox has previously conuted the vehicle       
        if found_box == True:
            continue

    #initialising time_value as difference between current and start time
    time_value = datetime.now() - startTime_video 
    #initialising seconds_value
    seconds_value = time_value.total_seconds()
    # calculating the number of cars detected per minute
    cars_passing_perMinute = vehicles_counter * 60 / seconds_value
    
    #Draw information on screen
    cv2.putText(frame_rate, f'Count:{vehicles_counter} in {int(seconds_value)}s - ({cars_passing_perMinute:.2f} cars/min)', (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0,0,0), 2, cv2.LINE_AA)
        
    #assign current_BoundBox to previous_BoundBox
    previous_BoundBox = current_BoundBox

    #Displaying the various frames on screen
    if check_rate == True:
        #displaying the webcam video_file
        cv2.imshow('Webcam', frame_rate)  
        #break playing the video if the key "q" is pressed or video is played completely 
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    
    else:
        break
        
#printing the required results
print(f'Total number of vehicles/cars passed: {vehicles_counter}')
print(f'Rate of vehicles/cars passing per minute: {cars_passing_perMinute:.2f}')

#Releasing the video_file object
video_file.release()

#destroying all the windows
cv2.destroyAllWindows()       
```
